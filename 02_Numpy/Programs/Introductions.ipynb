# Introduction to Data Analysis with NumPy

This short notebook explains why NumPy is a foundational library for data analysis in Python. It highlights core benefits, typical workflows, and recommended next steps for learners.

**Why NumPy?**
- Efficient numerical arrays implemented in C, providing speed and low memory usage.
- Powerful vectorized operations that avoid slow Python loops.
- Foundation for higher-level libraries like Pandas, SciPy, and scikit-learn.

**Quick Start**
1. Install NumPy: `pip install numpy`.
2. Import and create arrays: `import numpy as np; np.array([1,2,3])`.
3. Try basic ops: `arr + 1`, `arr.mean()`, `np.dot(a,b)`.

**Study tips**
- Run code interactively and inspect intermediate outputs.
- Compare NumPy operations with equivalent Python list code to see performance differences.
- Build small exercises: summary stats, masking, reshaping.
