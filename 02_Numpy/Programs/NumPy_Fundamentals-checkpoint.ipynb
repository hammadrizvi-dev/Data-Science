### NumPy is a Python library used for fast numerical computing. Its main data structure is the N-dimensional array called ndarray. NumPy arrays are generally faster and more memory-efficient than Python lists because they use homogeneous data, efficient memory storage, optimized low-level operations and vectorization.

### np.array() → Create array

### np.zeros() → Array of zeros

### np.ones() → Array of ones

### np.full() → Fill array with a value

### np.eye() → Identity matrix

### np.arange() → Values using a step

### np.linspace() → Evenly spaced values

### shape → Structure of array

size → Total elements

ndim → Number of dimensions

dtype → Data type

astype() → Change data type

reshape() → Change array shape

flatten() → Convert to 1D

Indexing → Select one element

Slicing → Select a range

Fancy Indexing → Select specific indexes

Boolean Masking → Filter using conditions

```python
import numpy as np 
```

```python
# Creating a array
arr = np.array([10,20,30,40])
print(arr)
```

```python
# 2d aray
mat = np.array([[10, 20, 30],[40, 50, 60]])
print(mat)
# 3d Array
# 2d aray
mate = np.array([[10, 20, 30],[40, 50, 60],[70,80,90]])
print(mate)
```

## Array from scratch


```python
zeros = np.zeros((3, 3))

print(zeros)
```

```python
ones = np.ones((2, 4))

print(ones)
```

```python
arr = np.full((2, 3), 7)

print(arr)
```

```python
identity = np.eye(3)

print(identity)
```

```python
numbers = np.arange(1, 10, 2)

print(numbers)
```

```python
numbers = np.linspace(0, 1, 5)

print(numbers)
```

## Array properties

```python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
```

```python
print(arr.shape)
```

```python
print(arr.size)
```

```python
print(arr.ndim)
```

```python
print(arr.dtype)
```

```python
print(arr.nbytes)
```

## Data type conversions

```python
arr = np.array([1.5, 2.8, 3.9])

print(arr.dtype)
```

```python
integer_array = arr.astype(np.int32)

print(integer_array)
```

## Reshaping

```python
arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)

print(reshaped)
```

## Flattening


```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

flat = matrix.flatten()

print(flat)
```

## Indexing

```python
arr = np.array([10, 20, 30, 40])

print(arr[0])
```

```python
print(arr[-1])
```

## Slicing


```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[1:4])
```

```python
print(arr[:3])
```

```python
print(arr[::2])
```
